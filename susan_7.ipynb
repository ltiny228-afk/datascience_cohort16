# Task 1 — Create Initial Data
* Create two variables using lists, representing product categories(Branch A & Branch B)

```python
branch_A = ["Iphones", "laptops", "Power_banks", "Ipads"]
branch_B = ["Shoes", "Bags", "Sunshades","Cloths"]
```

# Task 2 — Convert Lists to Sets
* Convert both lists to sets to remove duplicates.

```python
# Converting  lists to sets
set_A = set(branch_A)
set_B = set(branch_B)

print("Branch A categories:", set_A)
print("Branch B categories:", set_B)
```

# Task 3 — Modify the Sets
* Simulate new purchase updates:
* Add a new purchase category to Branch A
* Remove a category from Branch B

```python
set_A = {"Iphones", "laptops", "Power_banks", "Ipads"}
set_B = {"Shoes", "Bags", "Sunshades", "Cloths"}

# Adding new purchase category to Branch A
set_A.add("Smart_Watches")

# Removing a category from Branch B
set_B.remove("Cloths")

print("Updated Branch A categories:", set_A)
print("Updated Branch B categories:", set_B)
```

# Task 4 — Print a Final Report
* DATA SCIENCE SET ANALYSIS REPORT
--------------------------------
* Branch A Unique Categories: {'Rice', 'Oil', 'Bread', 'Beans'}
* Branch B Unique Categories: {'Milk', 'Oil', 'Bread', 'Beans'}
* Common Categories: {'Beans', 'Bread', 'Oil'}
* Only Branch A Purchased: {'Rice'}
* Only Branch B Purchased: {'Milk'}
* All Possible Categories Combined: {'Milk', 'Rice', 'Bread', 'Beans', 'Oil'}

```python
# New_sets
branch_A = {"Rice", "Oil", "Bread", "Beans"}
branch_B = {"Milk", "Oil", "Bread", "Beans"}

common_categories = branch_A.intersection(branch_B)
only_A = branch_A.difference(branch_B)
only_B = branch_B.difference(branch_A)
all_categories = branch_A.union(branch_B)

print("DATA SCIENCE SET ANALYSIS REPORT")
print("--------------------------------")
print("Branch A Unique Categories:", branch_A)
print("Branch B Unique Categories:", branch_B)
print("Common Categories:", common_categories)
print("Only Branch A Purchased:", only_A)
print("Only Branch B Purchased:", only_B)
print("All Possible Categories Combined:", all_categories)
```
