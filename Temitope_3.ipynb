```python
name = input("What is your name? ")
age = input("How old are you? ")
favorite_food = input("What is your favorite food? ")
location = input("Where do you live? ")
hobby = input("What is your hobby? ")

print("\n--- SURVEY SUMMARY ---")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Favorite Food: {favorite_food}")
print(f"Location: {location}")
print(f"Hobby: {hobby}")
```

```python
january = 4500
february = 5200
march = 6100
total_sales = january + february + march
average_sales = total_sales / 3

print("\n--- SALES REPORT ---")
print(f"Total Sales: {total_sales}")
print(f"Average Monthly Sales: {average_sales}")

if average_sales > 5000:
    print("Great quarter!")
else:
    print("Needs improvement!")
```

```python
raw_data = " data science is fun and powerful!!! "

cleaned = raw_data.strip()             
cleaned = cleaned.title()               
cleaned = cleaned.replace("Fun", "Interesting")

print("\nOriginal:", raw_data)
print("Cleaned:", cleaned)
```

```python
math = int(input("Enter Math score: "))
science = int(input("Enter Science score: "))
english = int(input("Enter English score: "))
total = math + science + english
average = total / 3

print("\n--- SCORE REPORT ---")
print(f"Total Score: {total}")
print(f"Average Score: {average}")

if average >= 70:
    print("Excellent")
elif average >= 50:
    print("Good")
else:
    print("Needs Improvement")
```
